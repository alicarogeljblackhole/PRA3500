```
"""
INSTRUCTIONS:
-for weathertoday_sunny_or_rainy variable insert either of the strings 'sunny' or 'rainy', depending on what is the weather like that day
-for forecast_for_how_many_days variable, specify for how many days in advance do you want to see the weather forecast (in task: 14)
-for how_many_repetitions variable, specify for how many repetitions you want to run the forecast program (the more the better for accuracy of predictions of course).
-without amandment to probabilities, set a=90, b=50, as given in the instructions
-meaning of variables:
  *a=if weather is sunny today, what are the odds that it will be sunny tomorrow? Give in %
  *b=if weather is rainy today, what are the odds that it will be sunny tomorrow? Give in %
"""

def weather_forecase(weathertoday_sunny_or_rainy, forecast_for_how_many_days, how_many_repetitions, a, b):
  #print logistics
  print("Starting weather: ", weathertoday_sunny_or_rainy)
  print("How many repetitions: ", how_many_repetitions)
  print("Weather Scheme:")
  print("If today is sunny, there is ", a, "% probability that tommorow will also be sunny.")
  print("If today is rainy, there is ", b, "% probability that tommorow will be sunny.")
  print("")

  #define sunny and rainy days lists
  sunnydays=[]
  rainydays=[]

  #make code suitable for an arbitrary abount of repetition
  for j in range(how_many_repetitions):
    state=[weathertoday_sunny_or_rainy] #weather forecast depends on the weather of the current day

  #make code suitable for predictions for an arbitrary amount of days in advance  
    for i in range(forecast_for_how_many_days - 1):
  #generate a random number between 1 and 100. Since the number is random, there is, for example, 90% probability that the number generated will be between 1 and 90 - 
  # - corresponding to 90% probability of the weather being sunny tommorow, in case it is sunny today, for instance (when a=90). 
      x=random.randint(1, 100)
      if state[i]=="sunny":
        if x<=a:
          state.append("sunny")
        else:
          state.append("rainy")
      elif state[i] == "rainy":
        if x<=b:
          state.append("sunny")
        else:
          state.append("rainy")

  #in case user fails to input today's weather correctly, here are the instructions printed
      else:
        print("Error. For variable weathertoday_sunny_or_rainy, only strings 'sunny' or 'rainy' are allowed.")

  #after each repetition, add the total number of sunny and rainy days respectively to the two lists created at the beginning
    sunnydays.append(state.count("sunny"))
    rainydays.append(state.count("rainy"))


  #how many sunny and rainy days did we predict? Some nice statistics, after all the required repetitions
  print("Sunny days list: ",sunnydays) #each element corresponds to the number of sunny days is each repetition
  print("Rainy days list:", rainydays)
  print("Portion rainy days: ", sum(rainydays)/(forecast_for_how_many_days*how_many_repetitions))
  print("Portion sunny days: ", sum(sunnydays)/(forecast_for_how_many_days*how_many_repetitions))
  print("""
  """)

  #make nice graphs
  #put n of repetitions on x-axis
  x_axis=list(range(1, (how_many_repetitions+1)))
  #graph it
  plt.bar(x_axis, rainydays, color="blue", label="Rainy Days")
  plt.bar(x_axis, sunnydays, bottom=rainydays, color='orange', label="Sunny Days")
  plt.ylabel("Days")
  plt.xlabel("Number of Repetition")
  plt.legend(loc="upper right")
  plt.title("Weather Forecast for 14 days (based on today's weather)")
  plt.show()


#let's test it!

print("standard, 1 repetition")
print("")
weather_forecase("sunny", 14, 1, 90, 50)
weather_forecase("rainy", 14, 1, 90, 50)
print("")
print("standard, 50 repetitions")
print("")
weather_forecase("sunny", 14, 50, 90, 50)
weather_forecase("rainy", 14, 50, 90, 50)
print("")
print("50 repetitions, play with a, b")
print("")
weather_forecase("sunny", 14, 50, 90, 90)
weather_forecase("rainy", 14, 50, 90, 90)
weather_forecase("sunny", 14, 50, 10, 20)
weather_forecase("rainy", 14, 50, 10, 20)
weather_forecase("sunny", 14, 50, 98, 2)
weather_forecase("rainy", 14, 50, 98, 2)

#plug&play!
```

## Observiations

It is evident that the weather scheme of some geographical significantly impacts the number of sunny and rainy days. This is very intuitive: if we set in the program for example that (1) if today is either sunny or rainy, tommorow will be also sunny with 90% certainty, and compare it to a case where (2) today's sunny weather results will result in tommorow's 90% rain/today's rain will result in 80% certain tommorrow's rain, then obviously in case (2) there will be many more rainy days than in case (1). Both cases are printed out above.

Here is another one illustrative example: suppose you live in the land of AngryBlackHoleMagic, where the Gods of General Relativity have jinxed the weather such that, if today is either sunny or rainy, tommorow will also be sunny/rainy respectively with 98% probability (last 2 graphs printed). As can be seen from the graphs, in such cases, once you start out with a rainy day you will probably not see the sun anymore, while if you start out with a sunny day all the plants on the land will die because of the severe lack of rain. Additionally, sometimes it still happens that even if you start out with a sunny/rainy day, a rainy/sunny day respectively will follow, but once the switch happens it is unlikely to happen again and retain the original weather condition. In either case, your plant will die (too much sun/too little rain). Lesson: stay away from AngryBlackHoleMagic, if you want your plants to survive.

This code will probably not be very accurate as weather forecasts are not and this code has also very limited functionality, but if you plug in even the wrong weather scheme, results will likely be extremely inaccurate. If you happen to live in extreme weather conditions - such the lands of GR Gods - then your predictions might be less accurate even. That is because on the short run, extreme values for a, b result in more unpredictable weather - you may, in AngryBlackHoleMagic land, based on today's rainy weather conclude that in the next 14 days it will be raining constantly, but maybe just the next day will be the one sunny day, which may ruin your 14-day-rain prediction completely, as then it will probably be sunny for the remaining 13 days as well.)

Extra: I mentioned above that weather forcasts are not very accurate. In fact, the weather is very sensible to extremely small change of initial conditions. That makes it very hard to predict. It is indeed deterministic, but it's prediction is practically impossible, because its beaviour is chaotic.
