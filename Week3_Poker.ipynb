Task: Create a deck of cards, shuffle the deck (randomize) and then draw 5 cards.


Create a check to see if you have any hands of cards which would satisfy the
criteria in the standard rules of poker.


Repeat the task sufficient times to generate statistics. Does your model follow the
expected pattern?

```
"""
*create a deck of cards
*create dictionary
-first digit in a range corresponds to hearts, diamonds, clubs and spades (1,2,3,4) respectively
-the second two digits correspond to the value of the card: 2,3,4,..., 10, 11 (=jack), 12 (=queen), 13 (=king), 14 (=ace)

"""
import random

#define hands lists
royalflush=[]
straightflush=[]
fourofakind=[]
fullhouse=[]
flush=[]
straight=[]
threeofakind=[]
twopair=[]
onepair=[]
highcard=[]
#myhighcard=[] #gets the values of your high cards, if you are interested in knowing them

valuesdictionary = {
  2: "Two",
  3: "Three",
  4: "Four",
  5: "Five",
  6: "Six",
  7: "Seven",
  8: "Eight",
  9: "Nine",
  10: "Ten",
  11: "Jack",
  12: "Queen",
  13: "King",
  14: "Ace"
}

colorsdictionary = {
    100: "Hearts",
    200: "Diamonds",
    300: "Clubs",
    400: "Spades"
}


"""
DEFINE A FUNCTION THAT WILL DRAW STATISTICS FROM POKER GAMES

*draw 5 cards at random
*get cards in order
*get info on which colors do you have (in order automatically) + range
*get info on which values do you have (in order) + range
*more to follow

ready-to-use notes:
-mycards=ordered list of 5 cards you have in your hands
-colors=ordered list of colors you have (spades=400, hearts=100, diamonds=200, clubs=300)
-values=ordered list of values you have (3, 7, 11 (=J), 12 (=Q), 13 (=K) for instance)
-originalvalues=same as values, just in the original order (so you can easily compose the card back from colors+values)
-rangecolors=range of colors (if you have a single color, 0)
-rangevalues=difference between the higest and lowest card you have (eg. if highest=13, and lowest=9, that would be 4 = flush)
    for ranges functions, if range=4, one sometimes needs to check if we have all the values within the range: -range4checker() function
-acecheck()= returns True if you have a combination 2,3,4,5,A, and False otherwise; 
  this exists because A can either be in combination with 2, 3,.. or with ..Q, K. The latter can be read off easily from rangevalues, 
  while the former possibility requred an extra function
-occurence_checker(n)= checks if any value in mycards appears exactly n times

"""


def pokerstatistics(repetitions):
  i=1
  while i<=repetitions:

    hearts=list(range(102, 115))
    diamonds=list(range(202, 215))
    clubs=list(range(302, 315))
    spades=list(range(402, 415))
    deck_of_cards=hearts+diamonds+clubs+spades

    mycards=[]
    choice1=random.choice(deck_of_cards)
    deck_of_cards.remove(choice1)
    choice2=random.choice(deck_of_cards)
    deck_of_cards.remove(choice2)
    choice3=random.choice(deck_of_cards)
    deck_of_cards.remove(choice3)
    choice4=random.choice(deck_of_cards)
    deck_of_cards.remove(choice4)
    choice5=random.choice(deck_of_cards)
    deck_of_cards.remove(choice5)
    mycards.append(choice1)
    mycards.append(choice2)
    mycards.append(choice3)
    mycards.append(choice4)
    mycards.append(choice5)

    #sort cards
    mycards=sorted(mycards)

    #get list of colors
    colors=[]
    for x in mycards:
      first_value=100*int(x/100)
      colors.append(first_value)

    #get range of colors
    rangecolors=colors[4]-colors[0]

    #get only the values out
    values=[mycards_i - colors_i for mycards_i, colors_i in zip(mycards, colors)]
    originalvalues=values
    values=sorted(values)

    #get range of values
    rangevalues=values[4]-values[0]

    #find if range is filled with all integers possible
    #ONLY use it for range 4 cases
    def range4checker():
      if rangevalues==4:
        if values[4]>values[3]>values[2]>values[1]>values[0]:
          return True
        else:
          return False
      else:
        pass

    #check for ace combination (ace, 2, 3, 4, 5 comination for instance; it should't go unmissed!)
    def acecheck():
      if rangecolors==0:
        if values[-1]==14 and values[0]==2 and values[1]==3 and values[2]==4:
          return True
        else:
          return False
      else:
        return False

    #checks if any value in mycards appears exactly n times
    def occurence_checker(n):
      a=[]
      for y in values:
        z=values.count(y)
        if z==n:
          a.append(True)
        else:
          pass
      if sum(a)>=1:
        return True
      else:
        return False

    #check how many pairs you have 
    #you ONLY run it after you've run occurence_checker(2)
    #(other frequency checks (for triplets etc) are trivial and unnecessary)
    def doubles_frequency(n_of_pairs):
      if occurence_checker(2)==True:
        b=[]
        for y in values:
          z=values.count(y)
          if z==2:
            b.append(True)
          else:
            pass
        if sum(b)==2*n_of_pairs:
          return True
        else:
          return False
      else:
        pass

    """
    Define functions for hands (aka 'rules for hands for poker')
    """

    #10, J, Q, K, A & same suit
    def royal_flush():
      if mycards==[110, 111, 112, 113, 114] or mycards==[210, 211, 212, 213, 214] or mycards==[310, 311, 312, 313, 314] or mycards==[410, 411, 412, 413, 414]:
        royalflush.append(True)   
      else:
        royalflush.append(False)

    #5 consequitive cards&same suit (except for royal flush)
    def straight_flush():
      if mycards!=[110, 111, 112, 113, 114] and mycards!=[210, 211, 212, 213, 214] and mycards!=[310, 311, 312, 313, 314] and mycards!=[410, 411, 412, 413, 414]:
        if rangecolors==0:
          if rangevalues==4 or acecheck()==True:
            straightflush.append(True)
          else:
            straightflush.append(False)
        else:
          straightflush.append(False)
      else:
        straightflush.append(False)

    #4 same values different colors, 1 whatever
    def four_of_a_kind():
      if values[0]==values[1]==values[2]==values[3] or values[4]==values[1]==values[2]==values[3]:
        fourofakind.append(True)
      else:
        fourofakind.append(False)
        
    #3 times one value plus 2 times another value
    def full_house():
      if occurence_checker(3)==True and occurence_checker(2)==True:
        fullhouse.append(True)
      else:
        fullhouse.append(False)

    #5 cards of the same color, not in numerical order
    def flush_flush():
      if rangevalues!=4 and acecheck()==False:
        if rangecolors==0:
          flush.append(True)
        else:
          flush.append(False)
      else:
        flush.append(False)

    #5 cards numerical order, not all are in the same suit
    def straight_straight():
      if rangecolors!=0:
        if rangevalues==4:
          if range4checker()==True:
            straight.append(True)
          else:
            straight.append(False)
        else:
          straight.append(False)
      else:
        straight.append(False)

    #3 times one value plus another value plus yet another value
    def three_of_a_kind():
      if occurence_checker(3)==True and occurence_checker(1)==True:
        threeofakind.append(True)
      else:
        threeofakind.append(False)

    #you have exactly 2 pairs
    def two_pair():
      if occurence_checker(2)==True:
        if doubles_frequency(2)==True:
          twopair.append(True)
        else:
          twopair.append(False)
      else:
        twopair.append(False)

    #you have exactly 1 pair
    def one_pair():
      if occurence_checker(3)==False and occurence_checker(2)==True:
        if doubles_frequency(1)==True: 
          onepair.append(True)
        else:
          onepair.append(False)
      else:
        onepair.append(False)

    #no matching cards
    def high_card():
      if occurence_checker(5)==False and occurence_checker(4)==False and occurence_checker(3)==False and occurence_checker(2)==False and rangevalues!=4 and acecheck()==False and rangecolors!=0:
        highcard.append(True)
        #myhighcard.append(values[-1])
      else:
        highcard.append(False)
        #myhighcard.append(False)

    #actual game
    """
    Create a check to see if you have any hands of cards which would satisfy the criteria in the standard rules of poker.

    """
    def runall():
      royal_flush()
      straight_flush()
      four_of_a_kind()
      full_house()
      flush_flush()
      straight_straight()
      three_of_a_kind()
      two_pair()
      one_pair()
      high_card()

    runall()

    i+=1
    if i>repetitions:
      break

  #find probabilities of getting each hand
  P_royalflush=sum(royalflush)/repetitions
  P_straightflush=sum(straightflush)/repetitions
  P_fourofakind=sum(fourofakind)/repetitions
  P_fullhouse=sum(fullhouse)/repetitions
  P_flush=sum(flush)/repetitions
  P_straight=sum(straight)/repetitions
  P_threeofakind=sum(threeofakind)/repetitions
  P_twopair=sum(twopair)/repetitions
  P_onepair=sum(onepair)/repetitions
  P_highcard=sum(highcard)/repetitions
  MyProbabilities=[P_royalflush, P_straightflush, P_fourofakind, P_fullhouse, P_flush, P_straight, P_threeofakind, P_twopair, P_onepair, P_highcard]

  if repetitions==1:
    yourvalues=[]
    yourcolors=[]
    for i in originalvalues:
      yourvalues.append(valuesdictionary[i])
    for j in colors:
      yourcolors.append(colorsdictionary[j])
    print("Your cards are: ")
    for k in range(5):
      print("* ", yourvalues[k], "of", yourcolors[k])

    if sum(royalflush)==1:
      print("Congratulations! Royal Flush!")
    if sum(straightflush)==1:
      print("Congratulations! Straight Flush!")
    if sum(fourofakind)==1:
      print("Congratulations! Four of a Kind!")
    if sum(fullhouse)==1:
      print("Congratulations! Full House!")
    if sum(flush)==1:
      print("Congratulations! Flush!")
    if sum(straight)==1:
      print("Congratulations! Straight!")
    if sum(threeofakind)==1:
      print("Congratulations! Three of a Kind!")
    if sum(twopair)==1:
      print("Congratulations! Two Pair!")
    if sum(onepair)==1:
      print("Congratulations! One Pair!")
    if sum(highcard)==1:
      print("Congratulations! High Card!")
    if sum(highcard)>=1:
      print("Your Highest Card is: ", valuesdictionary[values[-1]])
    else:
      pass
  else:
    print("Number of repetitions: ", repetitions)
    print("This code predicts:")
    print("Probability of getting Royal Flush is: ", MyProbabilities[0])
    print("Probability of getting Straight Flush is: ", MyProbabilities[1])
    print("Probability of getting Four Of A Kind is: ", MyProbabilities[2])
    print("Probability of getting Full House is: ", MyProbabilities[3])
    print("Probability of getting Flush is: ", MyProbabilities[4])
    print("Probability of getting Straight is: ", MyProbabilities[5])
    print("Probability of getting Three Of A Kind is: ", MyProbabilities[6])
    print("Probability of getting Two Pair is: ", MyProbabilities[7])
    print("Probability of getting One Pair is: ", MyProbabilities[8])
    print("Probability of getting High Card is: ", MyProbabilities[9])
    print("Found Probabilities list: ",MyProbabilities)
    print("")
    #expected pattern found online
    known_prob=[0.00000154,0.0000139,0.000240,0.001441,0.001965,0.003925,0.021128,0.047539,0.422569,0.501177]
    print("Known Probabilities list: ", known_prob)
    #comparison
    compare=[a_i - b_i for a_i, b_i in zip(MyProbabilities, known_prob)]
    comp=1-(sum([abs(number) for number in compare])/2)
    print("Compare my values with known values:")
    print("Likeliness: ", round(comp, 5))
    print("The closer the likeliness is to 1, the more my values match with the known ones.")
    print("")
    print("Known Probabilities data was obtained from: https://en.wikipedia.org/wiki/Poker_probability")

#get some results!
print("A single game:")
pokerstatistics(1)
print("")
print("Some Statistics, based on this code:")
pokerstatistics(10000000)
```

It can be seen that my probabilities match the know ones extremely well (99.9%!). That is based on 10 million repetitions. If the number of repetitions is very low (lower than approximately 50), then my probabilities would not match the known ones very well. That's easy to understand - the more repetitions, the less chance for error to significantly influence the results.
