```
# for inline plots in jupyter
%matplotlib inline
# import matplotlib
import matplotlib.pyplot as plt
# for latex equations
from IPython.display import Math, Latex
# for displaying images
from IPython.core.display import Image
```

```
# import seaborn
import seaborn as sns
# settings for seaborn plotting style
sns.set(color_codes=True)
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(5,5)})
```

```
from scipy.stats import norm
# generate random numbers from N(0,1)
data_normal = norm.rvs(size=10000,loc=0,scale=1)

from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=1000)

from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3, size=1000)

from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
```

```
def plot_graph(which_data):
  ax = sns.distplot(which_data,
                  bins=10,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
  ax.set(xlabel='title', ylabel='y axis')
```

```
plot_graph(data_normal)
```

```
plot_graph(data_expon)
```

```
plot_graph(data_poisson)
```

```
plot_graph(data_binom) 
```
